# 1d and 2d tight binding model 

## 1d case

The eigenvalues are given by $\epsilon_k=-2t \cos(k \cdot a)$. These eigenvalues constitute the quasi particle dispersion defined in the first Brillouin zone i.e. for $k\in \{-\pi/a , +\pi/2\}$. The following python script will plot this dispersion:

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
t = 1.0  # You can adjust this value
a = 1.0  # You can adjust this value

# Range for k: from -pi/a to pi/a
k = np.linspace(-np.pi/a, np.pi/a, 400)

# The function epsilon_k
epsilon_k = -2 * t * np.cos(k * a)

# Plotting the graph
plt.figure(figsize=(8, 4))
plt.plot(k, epsilon_k, label='ε_k = -2t cos(ka)')
plt.title('Plot of ε_k = -2t cos(ka)')
plt.xlabel('k')
plt.ylabel('ε_k')
plt.axhline(0, color='gray', lw=0.5)
plt.axvline(0, color='gray', lw=0.5)
plt.grid(True)
plt.legend()
plt.show()
```

## 2d case

We have now the dispersion $\epsilon_{\vec k}=-2t\sum_{i=1}^2\cos(k_i\cdot a)$, which we visualize in form of a contour plot.

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
t = 1.0  # You can adjust this value
a = 1.0  # You can adjust this value

# Range for k: from -pi/a to pi/a
k1 = np.linspace(-np.pi/a, np.pi/a, 400)
k2 = np.linspace(-np.pi/a, np.pi/a, 400)
K1, K2 = np.meshgrid(k1,k2)

#Define dispersion on 2d Brillouin zone
epsilon_k = -2 * t * (np.cos(K1 * a)+np.cos(K2 * a))
```

```python
contours = plt.contour(K1, K2, epsilon_k)
plt.clabel(contours, colors='black', fontsize=8, inline=True, fmt="%1.1f") 
plt.title('Plot of ε_k = -2t (cos(kx a)+cos(ky a))')
plt.xlabel('k_x')
plt.ylabel('k_y')
plt.show()
```
